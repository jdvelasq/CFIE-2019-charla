# Análisis: Matriz de autocorrelación

```python
import pandas as pd
from techminer import RecordsDataFrame

rdf = RecordsDataFrame(
    pd.read_json(
        'part-07.json', 
        orient='records', 
        lines=True))
```

## Authors

```python
rdf.auto_corr(
    column='Authors',
    sep=',',
    top_n=10
).heatmap(figsize=(8,8))
```

```python
rdf.auto_corr(
    column='Authors',
    sep=',',
    top_n=20
).heatmap(figsize=(10,10))
```

```python
rdf.auto_corr(
    column='Authors',
    sep=',',
    top_n=30
).heatmap(figsize=(12,12))
```

```python
rdf.auto_corr(
    column='Authors',
    sep=',',
    top_n=20
).map(figsize=(10,10))
```

```python
rdf.auto_corr(
    column='Authors',
    sep=',',
    top_n=20
).chord_diagram(figsize=(10,10))
```
